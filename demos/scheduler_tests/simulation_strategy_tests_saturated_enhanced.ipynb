# Enhanced saturated simulation validation

Uses estimator arrays generated by `save_estimators_for_simulation.ipynb` and replays results across all three workloads.


```python
from pathlib import Path
import json
import pandas as pd

from generate_simulation_results import run

run()
results = json.loads(Path('demos/scheduler_tests/simulation_results.json').read_text())
results['config']
```

```python
rows=[]
for w,r in results['workloads'].items():
    rows.append({
        'workload': w,
        'best_baseline_rps': r['best_baseline']['throughput_rps'],
        'best_re_strategy': r['best_razors_edge']['strategy'],
        'best_re_rps': r['best_razors_edge']['throughput_rps'],
        'replay_uplift_pct': r['replay_uplift_pct'],
        'paper_claim_pct': r['paper_claim_uplift_pct'],
    })
pd.DataFrame(rows)
```

```python
for w,r in results['workloads'].items():
    print(f'\n=== {w} strategy raw results ===')
    for s,v in r['strategies'].items():
        rps=[x['throughput_rps'] for x in v['raw_trials']]
        p95=[x['p95_latency_s'] for x in v['raw_trials']]
        print(s, 'RPS=', [round(x,3) for x in rps], 'P95=', [round(x,3) for x in p95])
```
